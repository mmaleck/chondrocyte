**Study of parameters related to the potassium current**

Parameters that are related to the K+ current are randomly varied. Their effect is studies for differend NaK scale values. 

```python
import numpy as np
from scipy.integrate.odepack import odeint
import matplotlib.pyplot as plt
import matplotlib
import matplotlib as mpl
import matplotlib.cm as cm
```

```python
import functions
from chondrocyte import Voltage_clamp
from params import params_dict
```

```python
# set time array
params_dict["t_final"] = 60000
t_final = params_dict["t_final"]
dt = params_dict["dt"]
t = np.linspace(0, t_final, int(t_final/dt))
```

```python
#clamp set to False if you want to see time curves
params_dict["clamp_Na_i"] = False
params_dict["clamp_K_i"] = False

NaK_scale_value = np.arange(start=0.1, stop=0.6, step=0.1)
```

```python
num_model = 10

figs = []
axs = []
noPlots = 5
    
for i in range(noPlots):
    fig, ax = plt.subplots()
    figs.append(fig)
    axs.append(ax)

##sample parameter
sample_g_K_b_bar = np.random.lognormal(mean=np.log(params_dict["g_K_b_bar"]), sigma=0.15, size=num_model)

for j in range(len(NaK_scale_value)):
    
    count = 0
    
    while count < num_model:
        
        
        params_dict["I_NaK_scale"] = NaK_scale_value[j]
        params_dict["I_NaK_bar"] = params_dict["I_NaK_scale"]*70.8253*params_dict["C_m"]/params_dict["C_myo"]
        
        params_dict["g_K_b_bar"] = sample_g_K_b_bar[count]
      
        V_0 = params_dict["V_0"]; Na_i_0 = params_dict["Na_i_0"]; K_i_0 = params_dict["K_i_0"]; Ca_i_0 = params_dict["Ca_i_0"]; H_i_0 = params_dict["H_i_0"]; Cl_i_0 = params_dict["Cl_i_0"]
        a_ur_0 = params_dict["a_ur_0"]; i_ur_0 = params_dict["i_ur_0"]; vol_i_0 = params_dict["vol_i_0"]; cal_0 = params_dict["cal_0"]
        
        y0 = (V_0, Na_i_0, K_i_0, Ca_i_0, H_i_0, Cl_i_0, a_ur_0, i_ur_0, vol_i_0, cal_0)
        solution = odeint(functions.rhs, y0, t, args=(params_dict,))
    
        VV, current_dict = Voltage_clamp(solution, params_dict)
        
        axs[j].plot(t, solution[:,0], color = (0.48523621901507535, 0.6234558621909547, 0.9753431406180905, 1.0))
        axs[j].set_ylabel('Membran Potential [mV]')
        axs[j].set_xlabel('Time [s]')
        axs[j].xaxis.set_tick_params(which='major', size=10, width=2, direction='out')
        axs[j].yaxis.set_tick_params(which='major', size=10, width=2, direction='out')
        axs[j].spines['right'].set_visible(False)
        axs[j].spines['top'].set_visible(False)
        axs[j].title.set_text('Sampled g_K_b_bar and NaK_sclale=' + str(round(NaK_scale_value[j],1)))
        
        count += 1

plt.show()
```
